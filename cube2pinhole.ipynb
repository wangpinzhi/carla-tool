```python
import torch
import numpy as np
import math
import cv2
import torch.nn.functional as F


cubeW = 1080
outW = 1280
fov = 100

outS = math.ceil(outW/(2*math.tan(fov/360*math.pi)))
cube_front = cv2.imread(r'output\huawei_demo\cubemap\cm_rgb9_up_11768.png')
cube_front = cube_front.transpose(2,0,1)
cube_front = np.expand_dims(cube_front, axis=0)


cube_front = torch.from_numpy(cube_front)
cube_front = cube_front.type(dtype=torch.float32)
```

```python
if outS*2 < outW:
    up_grid  = torch.zeros((1,outW//2-outS,outW,2),dtype=torch.float32)
    
    for i in range(outW//2-outS):
        for j in range(outW):
            print(j-outW//2)
            up_grid[0][i][j][0] = (cubeW*(j-outW//2)/(outW-2*i))/cubeW*2
            up_grid[0][i][j][1] = (cubeW*outS/(outW-2*i))/cubeW*2
            

    out = F.grid_sample(cube_front, up_grid, mode='bilinear', padding_mode='zeros', align_corners=False)


    out = out.cpu().numpy()
    out = out.astype(np.uint8)
    out = out.squeeze(axis=0)
    out = out.transpose(1,2,0)

    cv2.imshow('test', out)
    cv2.waitKey(-1)
    print(out.shape)

```
